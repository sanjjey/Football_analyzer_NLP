# PREDICTION FOR REAL MADRID VS BARCELONA

```python
import json
import os
import re
import time
import pandas as pd
import numpy as np
import nltk
import spacy
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from mplsoccer import Pitch
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from groq import Groq
from jsonschema import validate, ValidationError
from collections import defaultdict

# ─────────────────────────────────────────────────────────────────────────────
# DEPENDENCIES: pip install groq spacy mplsoccer jsonschema nltk pandas numpy
#               python -m spacy download en_core_web_md
# ─────────────────────────────────────────────────────────────────────────────

try:
    nltk.data.find('sentiment/vader_lexicon.zip')
except LookupError:
    nltk.download('vader_lexicon', quiet=True)

# ─────────────────────────────────────────────────────────────────────────────
# CONFIGURATION
# ─────────────────────────────────────────────────────────────────────────────
API_KEY     ="<api_key>"
client      = Groq(api_key=API_KEY)
GROQ_MODEL  = "llama-3.3-70b-versatile"

# ─────────────────────────────────────────────────────────────────────────────
# JSON SCHEMAS FOR STRICT VALIDATION
# ─────────────────────────────────────────────────────────────────────────────
NARRATIVE_SCHEMA = {
    "type": "object",
    "required": ["play_by_play", "new_scorers", "fouls_mentioned",
                 "cards_mentioned", "topic_label", "crowd_intensity",
                 "referee_involvement"],
    "properties": {
        "play_by_play":         {"type": "string", "minLength": 30},
        "new_scorers":          {"type": "array",  "items": {"type": "string"}},
        "fouls_mentioned":      {"type": "integer","minimum": 0, "maximum": 50},
        "cards_mentioned":      {"type": "integer","minimum": 0, "maximum": 20},
        "topic_label":          {"type": "string",
                                 "enum": ["tactical","emotional","referee","performance"]},
        "crowd_intensity":      {"type": "number", "minimum": 0.0, "maximum": 1.0},
        "referee_involvement":  {"type": "number", "minimum": 0.0, "maximum": 1.0},
    },
    "additionalProperties": True,
}

COACH_SCHEMA = {
    "type": "object",
    "required": ["home_team_analysis", "away_team_analysis"],
    "properties": {
        "home_team_analysis": {
            "type": "object",
            "required": ["substitution_alert", "tactical_shift", "coach_recommendation"],
            "properties": {
                "substitution_alert":    {"type": ["string", "null"]},
                "tactical_shift":        {"type": "string", "minLength": 5},
                "coach_recommendation":  {"type": "string", "minLength": 10},
            }
        },
        "away_team_analysis": {
            "type": "object",
            "required": ["substitution_alert", "tactical_shift", "coach_recommendation"],
            "properties": {
                "substitution_alert":    {"type": ["string", "null"]},
                "tactical_shift":        {"type": "string", "minLength": 5},
                "coach_recommendation":  {"type": "string", "minLength": 10},
            }
        }
    }
}

VALIDATION_SCHEMA = {
    "type": "object",
    "required": ["consistency_score", "explanation"],
    "properties": {
        "consistency_score": {"type": "number", "minimum": 0.0, "maximum": 1.0},
        "explanation":       {"type": "string", "minLength": 10},
    }
}

SCORESTATE_RE = re.compile(r"SCORESTATE:\s*.*?(\d+)\s*-\s*(\d+)", re.IGNORECASE)

# ─────────────────────────────────────────────────────────────────────────────
# ██  VERIFICATION ENGINE
# ─────────────────────────────────────────────────────────────────────────────
class LLMOutputVerifier:
    """
    Runs multi-dimensional quality checks on every LLM output.

    Metric families
    ───────────────
    1. Schema Compliance       – JSON structure & type validation
    2. Field Completeness      – Required keys present & non-empty
    3. Semantic Coherence      – SCORESTATE present, topic valid, etc.
    4. Score Integrity         – Non-decreasing, realistic jumps
    5. Hallucination Proxy     – Ungrounded goal spikes, implausible values
    6. Sentiment Calibration   – VADER vs LLM crowd-intensity correlation
    7. Latency Tracking        – API response time
    8. Retry / Fallback Rate   – How often the LLM fails
    9. Cross-Segment Drift     – Topic & narrative consistency over time
    10. Coaching Specificity   – Tactical advice richness score
    """

    def __init__(self):
        self.log: list[dict] = []                # Per-call records
        self.segment_history: list[dict] = []    # Cross-segment state
        self.summary_stats: dict = defaultdict(list)

    # ── helpers ─────────────────────────────────────────────────────────────
    @staticmethod
    def _word_count(text: str) -> int:
        return len(text.split()) if text else 0

    @staticmethod
    def _schema_check(data: dict, schema: dict) -> tuple[bool, str]:
        try:
            validate(instance=data, schema=schema)
            return True, "OK"
        except ValidationError as e:
            return False, e.message

    # ── METRIC 1: Schema Compliance ─────────────────────────────────────────
    def check_schema(self, data: dict, schema: dict, call_name: str) -> dict:
        passed, msg = self._schema_check(data, schema)
        return {"metric": "schema_compliance", "call": call_name,
                "passed": passed, "detail": msg}

    # ── METRIC 2: Field Completeness ────────────────────────────────────────
    def check_completeness(self, data: dict, required_keys: list,
                           call_name: str) -> dict:
        missing, empty = [], []
        for k in required_keys:
            if k not in data:
                missing.append(k)
            elif data[k] in (None, "", [], {}):
                empty.append(k)
        passed = not missing and not empty
        detail = (f"Missing: {missing}; Empty: {empty}") if not passed else "All fields present"
        return {"metric": "field_completeness", "call": call_name,
                "passed": passed, "detail": detail,
                "missing_keys": missing, "empty_keys": empty}

    # ── METRIC 3: Semantic Coherence ────────────────────────────────────────
    def check_narrative_coherence(self, narrative_data: dict,
                                  time_label: str) -> dict:
        pbp       = narrative_data.get("play_by_play", "")
        has_score = bool(SCORESTATE_RE.search(pbp))
        wc        = self._word_count(pbp)
        topic_ok  = narrative_data.get("topic_label") in \
                    ["tactical", "emotional", "referee", "performance"]

        issues = []
        if not has_score:  issues.append("SCORESTATE tag missing from play_by_play")
        if wc < 40:        issues.append(f"play_by_play too short ({wc} words)")
        if not topic_ok:   issues.append(f"Invalid topic_label: {narrative_data.get('topic_label')}")

        return {"metric": "semantic_coherence", "segment": time_label,
                "passed": not issues, "scorestate_found": has_score,
                "word_count": wc, "topic_valid": topic_ok, "issues": issues}

    # ── METRIC 4: Score Integrity ────────────────────────────────────────────
    def check_score_integrity(self, prev_h: int, prev_a: int,
                              new_h: int, new_a: int,
                              time_label: str) -> dict:
        issues = []
        if new_h < prev_h:
            issues.append(f"Home score decreased: {prev_h} → {new_h}")
        if new_a < prev_a:
            issues.append(f"Away score decreased: {prev_a} → {new_a}")
        delta_h = new_h - prev_h
        delta_a = new_a - prev_a
        if delta_h > 4:
            issues.append(f"Unrealistic home goal burst (+{delta_h} in 30 min)")
        if delta_a > 4:
            issues.append(f"Unrealistic away goal burst (+{delta_a} in 30 min)")
        if new_h > 15 or new_a > 15:
            issues.append(f"Implausible total score: {new_h}-{new_a}")

        return {"metric": "score_integrity", "segment": time_label,
                "passed": not issues, "prev_score": f"{prev_h}-{prev_a}",
                "new_score": f"{new_h}-{new_a}",
                "delta_home": delta_h, "delta_away": delta_a, "issues": issues}

    # ── METRIC 5: Hallucination Proxy ───────────────────────────────────────
    def check_hallucination_proxy(self, narrative_data: dict,
                                  raw_text: str, time_label: str) -> dict:
        """
        Flags potential hallucinations by checking if claimed scorers
        appear anywhere in the raw transcript.
        """
        scorers   = narrative_data.get("new_scorers", [])
        raw_lower = raw_text.lower()
        phantom   = [s for s in scorers
                     if s and s.lower() not in raw_lower and s not in ("None","")]
        ratio     = len(phantom) / max(len(scorers), 1)
        passed    = ratio < 0.5  # Allow < 50% unverified names

        return {"metric": "hallucination_proxy", "segment": time_label,
                "passed": passed, "claimed_scorers": scorers,
                "unverified_scorers": phantom,
                "phantom_ratio": round(ratio, 3)}

    # ── METRIC 6: Sentiment Calibration ─────────────────────────────────────
    def check_sentiment_calibration(self, vader_compound: float,
                                    llm_crowd_intensity: float,
                                    time_label: str) -> dict:
        """
        VADER compound ∈ [-1,1]; crowd_intensity ∈ [0,1].
        They should positively correlate. We flag if they diverge sharply.
        """
        normalised_vader = (vader_compound + 1) / 2   # → [0,1]
        gap              = abs(normalised_vader - llm_crowd_intensity)
        passed           = gap < 0.45

        return {"metric": "sentiment_calibration", "segment": time_label,
                "passed": passed,
                "vader_compound": round(vader_compound, 4),
                "llm_crowd_intensity": round(llm_crowd_intensity, 4),
                "normalised_vader": round(normalised_vader, 4),
                "calibration_gap": round(gap, 4),
                "detail": "OK" if passed else f"Large gap ({gap:.2f}) – LLM may be over/under-stating crowd"}

    # ── METRIC 7: Latency Tracking ───────────────────────────────────────────
    def record_latency(self, call_name: str, elapsed_s: float) -> dict:
        return {"metric": "api_latency_s", "call": call_name,
                "elapsed": round(elapsed_s, 3),
                "slow": elapsed_s > 15}

    # ── METRIC 8: Retry / Fallback Rate ─────────────────────────────────────
    def record_call_result(self, call_name: str, success: bool,
                           fallback_used: bool) -> dict:
        self.summary_stats["successes"].append(int(success))
        self.summary_stats["fallbacks"].append(int(fallback_used))
        return {"metric": "call_success", "call": call_name,
                "success": success, "fallback_used": fallback_used}

    # ── METRIC 9: Cross-Segment Drift ───────────────────────────────────────
    def check_cross_segment_drift(self, topic: str, sentiment: float,
                                  time_label: str) -> dict:
        """Detects abrupt, suspicious topic flips between segments."""
        result = {"metric": "cross_segment_drift", "segment": time_label,
                  "topic": topic, "sentiment": round(sentiment, 4),
                  "passed": True, "detail": "First segment"}

        if self.segment_history:
            prev = self.segment_history[-1]
            topic_flip = (prev["topic"] != topic)
            sent_jump  = abs(sentiment - prev["sentiment"])

            suspicious = topic_flip and sent_jump > 0.7
            result.update({
                "prev_topic": prev["topic"], "topic_flipped": topic_flip,
                "sentiment_jump": round(sent_jump, 4),
                "passed": not suspicious,
                "detail": ("Suspicious abrupt topic+sentiment swing"
                           if suspicious else "Drift within acceptable range")
            })

        self.segment_history.append({"topic": topic, "sentiment": sentiment})
        return result

    # ── METRIC 10: Coaching Specificity ─────────────────────────────────────
    def check_coaching_specificity(self, coach_data: dict,
                                   time_label: str) -> dict:
        """
        Scores coaching output richness:
        – Uses a named player for sub alert (+2)
        – Tactical shift has a formation number e.g. 4-3-3 (+2)
        – Recommendation ≥ 15 words (+1)
        """
        scores, details = {}, {}
        for side in ("home_team_analysis", "away_team_analysis"):
            d     = coach_data.get(side, {})
            score = 0
            sub   = d.get("substitution_alert", "") or ""
            tact  = d.get("tactical_shift", "") or ""
            rec   = d.get("coach_recommendation", "") or ""

            if sub.lower() not in ("none", "n/a", "", "null") and len(sub) > 2:
                score += 2
            if re.search(r'\d-\d-\d', tact):
                score += 2
            if self._word_count(rec) >= 15:
                score += 1

            scores[side]  = score
            details[side] = {"sub_alert_named": score >= 2,
                             "formation_cited": score >= 4,
                             "rec_length_ok": self._word_count(rec) >= 15}

        total   = sum(scores.values())
        passed  = total >= 3  # At least half the possible 10 points
        return {"metric": "coaching_specificity", "segment": time_label,
                "passed": passed, "total_score": total,
                "max_score": 10, "breakdown": details}

    # ── METRIC 11: Consistency Score Sanity ─────────────────────────────────
    def check_validation_output(self, val_data: dict, time_label: str) -> dict:
        c   = val_data.get("consistency_score", -1)
        exp = val_data.get("explanation", "")
        issues = []
        if not (0.0 <= c <= 1.0):
            issues.append(f"consistency_score out of range: {c}")
        if self._word_count(exp) < 5:
            issues.append("Explanation too brief")
        return {"metric": "validation_output_sanity", "segment": time_label,
                "passed": not issues, "consistency_score": c,
                "explanation_word_count": self._word_count(exp),
                "issues": issues}

    # ── Log & append ────────────────────────────────────────────────────────
    def log_check(self, result: dict):
        self.log.append(result)
        status = "✅ PASS" if result.get("passed", True) else "❌ FAIL"
        print(f"   [{status}] {result['metric']}"
              + (f" | {result.get('detail','')}" if result.get('detail') else "")
              + (f" | issues: {result.get('issues','')}" if result.get('issues') else ""))

    # ── Final Report ─────────────────────────────────────────────────────────
    def print_summary_report(self):
        print("\n" + "═"*80)
        print("  🔬  LLM OUTPUT VERIFICATION SUMMARY REPORT")
        print("═"*80)

        total  = len(self.log)
        passed = sum(1 for r in self.log if r.get("passed", True))
        failed = total - passed

        print(f"\n  Total checks   : {total}")
        print(f"  ✅ Passed       : {passed}  ({passed/max(total,1)*100:.1f}%)")
        print(f"  ❌ Failed       : {failed}  ({failed/max(total,1)*100:.1f}%)")

        # Group by metric family
        family_map: dict[str, list] = defaultdict(list)
        for r in self.log:
            family_map[r["metric"]].append(r)

        print("\n  ─── Per-Metric Breakdown ───")
        for family, records in family_map.items():
            p = sum(1 for r in records if r.get("passed", True))
            print(f"  {family:<35} {p}/{len(records)} passed")

        # Latency stats
        latencies = [r["elapsed"] for r in self.log if r.get("metric") == "api_latency_s"]
        if latencies:
            print(f"\n  ─── Latency Stats ───")
            print(f"  Mean  : {np.mean(latencies):.2f}s")
            print(f"  Max   : {np.max(latencies):.2f}s")
            print(f"  Slow calls (>15s): {sum(1 for l in latencies if l>15)}")

        # Retry / Fallback
        fb = self.summary_stats.get("fallbacks", [])
        if fb:
            print(f"\n  ─── Fallback Rate ───")
            print(f"  Fallback used: {sum(fb)}/{len(fb)} calls ({sum(fb)/len(fb)*100:.1f}%)")

        # Failed checks detail
        failures = [r for r in self.log if not r.get("passed", True)]
        if failures:
            print("\n  ─── Failed Checks Detail ───")
            for f in failures:
                seg  = f.get("segment", f.get("call", "?"))
                issl = f.get("issues", f.get("detail", ""))
                print(f"  ● [{f['metric']}] @ {seg}: {issl}")

        print("\n" + "═"*80 + "\n")

    # ── Export ───────────────────────────────────────────────────────────────
    def export_log_csv(self, path: str = "verification_log.csv"):
        pd.DataFrame(self.log).to_csv(path, index=False)
        print(f"  Verification log exported → {path}")


# ─────────────────────────────────────────────────────────────────────────────
# ██  COMBINED PIPELINE
# ─────────────────────────────────────────────────────────────────────────────
class SoccerCombinedPipeline:

    def __init__(self, h1_path: str, h2_path: str):
        self.h1_path = h1_path
        self.h2_path = h2_path

        print("Loading NLP Models (VADER & spaCy)...")
        self.sia = SentimentIntensityAnalyzer()
        self.nlp = spacy.load("en_core_web_md")

        self.teams: list[str] = []
        self.dataset: list[dict] = []
        self.verifier = LLMOutputVerifier()

    # ─────────────────────────────────────────────────────────────────────────
    # STEP 1: DATA INGESTION
    # ─────────────────────────────────────────────────────────────────────────
    def load_and_combine(self) -> list[dict]:
        def get_segs(path):
            if not os.path.exists(path):
                raise FileNotFoundError(f"File not found: {path}")
            with open(path, 'r', encoding='utf-8') as f:
                data = json.load(f)
            segs = data.get("segments", {})
            return sorted(segs.items(), key=lambda x: int(x[0]))

        h1_segs = get_segs(self.h1_path)
        h2_segs = get_segs(self.h2_path)

        combined = [{"start": v[0], "text": v[2].replace('\n', ' ').strip()}
                    for _, v in h1_segs]
        combined += [{"start": v[0] + 2700, "text": v[2].replace('\n', ' ').strip()}
                     for _, v in h2_segs]
        return combined

    def identify_teams(self, combined_data: list[dict]):
        print("Extracting team names via Groq LLM...")
        intro  = " ".join(s['text'] for s in combined_data if s['start'] <= 600)
        prompt = (f"Identify the TWO competing football clubs.\n"
                  f"Return JSON: {{\"home_team\": \"<String>\", \"away_team\": \"<String>\"}}\n"
                  f"Transcript: \"{intro}\"")

        fallback_used = False
        t0 = time.time()
        try:
            resp = client.chat.completions.create(
                messages=[{"role": "user", "content": prompt}],
                model=GROQ_MODEL,
                response_format={"type": "json_object"},
            )
            data = json.loads(resp.choices[0].message.content)
            self.teams = [data.get("home_team", "Home Team"),
                          data.get("away_team",  "Away Team")]
        except Exception as e:
            print(f"  Team extraction failed: {e}")
            self.teams = ["Home Team", "Away Team"]
            fallback_used = True

        elapsed = time.time() - t0
        self.verifier.log_check(self.verifier.record_latency("identify_teams", elapsed))
        self.verifier.log_check(
            self.verifier.record_call_result("identify_teams", not fallback_used, fallback_used))
        print(f"  Match: {self.teams[0]} (Home) vs {self.teams[1]} (Away)")
    def compress_transcript(self, raw_text: str) -> str:
        """
        Compresses raw ASR transcript by filtering out neutral, non-action sentences.
        Retains sentences with high emotional sentiment or key tactical football terms.
        """
        # Fast sentence splitting using regex (looks for periods, exclamation marks, etc.)
        sentences = re.split(r'(?<=[.!?])\s+', raw_text)

        # High-value vocabulary for soccer analytics
        keywords = {
            'goal', 'score', 'shot', 'save', 'foul', 'card', 'yellow',
            'red', 'penalty', 'corner', 'sub', 'replace', 'miss',
            'offside', 'var', 'cross', 'header', 'tackle', 'press', 'injury'
        }

        compressed_sentences = []

        for sent in sentences:
            if not sent.strip():
                continue

            sent_lower = sent.lower()

            # Rule 1: Does it contain a tactical keyword?
            has_keyword = any(kw in sent_lower for kw in keywords)

            # Rule 2: Is the commentator highly emotional? (VADER compound score)
            # Threshold of 0.4 catches strong excitement or sharp disappointment
            sentiment_score = self.sia.polarity_scores(sent)['compound']
            is_emotional = abs(sentiment_score) > 0.4

            # Keep the sentence if it passes either rule
            if has_keyword or is_emotional:
                compressed_sentences.append(sent.strip())

        compressed_text = " ".join(compressed_sentences)

        # Calculate and print compression stats
        orig_words = len(raw_text.split())
        comp_words = len(compressed_text.split())
        ratio = (1 - (comp_words / max(orig_words, 1))) * 100

        print(f"  ✂️ [COMPRESSION] Reduced from {orig_words} to {comp_words} words (-{ratio:.1f}%)")

        # Safety Fallback: If the filter was too aggressive and wiped out the text,
        # revert to the original raw text to prevent pipeline failure.
        if comp_words < 30 and orig_words > 100:
            print("  ⚠️ [COMPRESSION] Warning: Too aggressive. Bypassing compression.")
            return raw_text

        return compressed_text

    # ─────────────────────────────────────────────────────────────────────────
    # STEP 2a: NARRATIVE GENERATION (from Pipeline 1)
    # ─────────────────────────────────────────────────────────────────────────
    def generate_narrative(self, raw_text: str, t1: str, t2: str,
                           cur_h: int, cur_a: int) -> tuple[dict, bool, float]:
        prompt = f"""
        You are a sports journalist. Read this 30-min transcript between {t1} (Home) and {t2} (Away).
        Write a chronological play_by_play summary. Track goals, near misses, fouls, and cards.
        MANDATORY ENDING: SCORESTATE: {t1} X - Y {t2}
        (Replace X,Y with cumulative integer scores, adding to previous {cur_h}-{cur_a})

        Return JSON:
        {{
            "play_by_play": "<summary ending with SCORESTATE>",
            "new_scorers": ["<string>"],
            "fouls_mentioned": 0, "cards_mentioned": 0,
            "topic_label": "tactical",
            "crowd_intensity": 0.5, "referee_involvement": 0.5
        }}
        Transcript: "{raw_text}"
        """
        fallback_used = False
        t0 = time.time()
        try:
            resp = client.chat.completions.create(
                messages=[{"role": "user", "content": prompt}],
                model=GROQ_MODEL,
                response_format={"type": "json_object"},
            )
            data = json.loads(resp.choices[0].message.content)
        except Exception as e:
            print(f"  Narrative API Error: {e}")
            data = {
                "play_by_play": f"Error processing segment. SCORESTATE: {t1} {cur_h} - {cur_a} {t2}",
                "new_scorers": [], "fouls_mentioned": 0, "cards_mentioned": 0,
                "topic_label": "performance", "crowd_intensity": 0.5, "referee_involvement": 0.5
            }
            fallback_used = True
        return data, fallback_used, time.time() - t0

    # ─────────────────────────────────────────────────────────────────────────
    # STEP 2b: COACHING ADVICE (from Pipeline 2)
    # ─────────────────────────────────────────────────────────────────────────
    def get_coach_advice(self, narrative: str, home: str, away: str) -> tuple[dict, bool, float]:
        prompt = f"""
        You are an elite football manager. Analyze this 30-min match segment:
        "{narrative}"

        Provide tactical recommendations and substitution alerts for {home} and {away}.
        Identify specific players making errors or showing fatigue where mentioned.

        Return strictly JSON:
        {{
            "home_team_analysis": {{
                "substitution_alert": "<player name or null>",
                "tactical_shift": "<e.g. Change to 4-5-1>",
                "coach_recommendation": "<brief strategic advice>"
            }},
            "away_team_analysis": {{
                "substitution_alert": "<player name or null>",
                "tactical_shift": "<e.g. High press required>",
                "coach_recommendation": "<brief strategic advice>"
            }}
        }}
        """
        fallback_used = False
        t0 = time.time()
        try:
            resp = client.chat.completions.create(
                messages=[{"role": "user", "content": prompt}],
                model=GROQ_MODEL,
                response_format={"type": "json_object"},
            )
            data = json.loads(resp.choices[0].message.content)
        except Exception as e:
            print(f"  Coaching API Error: {e}")
            data = {"home_team_analysis": {"substitution_alert": None,
                                           "tactical_shift": "N/A",
                                           "coach_recommendation": "N/A"},
                    "away_team_analysis": {"substitution_alert": None,
                                           "tactical_shift": "N/A",
                                           "coach_recommendation": "N/A"}}
            fallback_used = True
        return data, fallback_used, time.time() - t0

    # ─────────────────────────────────────────────────────────────────────────
    # STEP 2c: PREDICTION VALIDATION (from Pipeline 1)
    # ─────────────────────────────────────────────────────────────────────────
    def validate_prediction(self, features: dict, prediction: str,
                            win_prob: float) -> tuple[dict, bool, float]:
        prompt = f"""
        Given these 30-min football segment features:
        - Topic: {features['topic_label']}
        - Sentiment Label: {features['sentiment_label']}
        - Momentum Shift: {features['momentum_label']}
        - Tactical State: {features['tactical_state']}
        - Events: {features['goals_diff']} Goal Difference, {features['fouls']} Fouls, {features['cards']} Cards

        An XGBoost model predicted: {prediction} (Home Win Probability: {win_prob:.2f}).
        Is this prediction consistent with the narrative context?

        Return JSON:
        {{
            "consistency_score": 0.0,
            "explanation": "<brief explanation>"
        }}
        """
        fallback_used = False
        t0 = time.time()
        try:
            resp = client.chat.completions.create(
                messages=[{"role": "user", "content": prompt}],
                model=GROQ_MODEL,
                response_format={"type": "json_object"},
            )
            data = json.loads(resp.choices[0].message.content)
        except Exception as e:
            print(f"  Validation API Error: {e}")
            data = {"consistency_score": 0.5, "explanation": "Validation failed."}
            fallback_used = True
        return data, fallback_used, time.time() - t0

    # ─────────────────────────────────────────────────────────────────────────
    # STEP 3: DETERMINISTIC NLP EXTRACTION
    # ─────────────────────────────────────────────────────────────────────────
    def extract_nlp_features(self, text: str,
                             fallback_h: int, fallback_a: int) -> tuple[float, int, int]:
        sentiment = self.sia.polarity_scores(text)['compound']
        match     = SCORESTATE_RE.search(text)
        if match:
            return sentiment, int(match.group(1)), int(match.group(2))
        return sentiment, fallback_h, fallback_a

    # ─────────────────────────────────────────────────────────────────────────
    # STEP 4: CATEGORICAL LABELS
    # ─────────────────────────────────────────────────────────────────────────
    @staticmethod
    def categorize_sentiment(s: float) -> str:
        if s >= 0.5:  return "Highly Exciting"
        if s >= 0.05: return "Positive / Active"
        if s > -0.05: return "Neutral / Balanced"
        return "Tense / Negative"

    @staticmethod
    def categorize_intensity(s: float) -> str:
        if s >= 0.7: return "Roaring / Electric"
        if s >= 0.4: return "Moderate / Murmuring"
        return "Quiet"

    @staticmethod
    def categorize_referee(s: float) -> str:
        if s >= 0.6: return "Strict / Highly Involved"
        if s >= 0.3: return "Moderate"
        return "Letting the Game Flow"

    @staticmethod
    def categorize_momentum(delta: float) -> str:
        if delta > 0.3:  return "Home Surge"
        if delta < -0.3: return "Away Surge"
        return "Stable"

    @staticmethod
    def categorize_tactics(sim: float) -> str:
        if sim == 0.0: return "N/A (Match Start)"
        if sim > 0.85: return "Stagnant / Consistent"
        return "Major Tactical Shift"

    # ─────────────────────────────────────────────────────────────────────────
    # STEP 5: SPATIAL HEATMAP (from Pipeline 2)
    # ─────────────────────────────────────────────────────────────────────────
    def generate_heatmap(self, possession_home: float, sentiment: float,
                         time_label: str):
        pitch = Pitch(pitch_type='statsbomb', pitch_color='#1a1a1a', line_color='#c7d5cc')
        fig, ax = pitch.draw(figsize=(10, 6))

        mean_x  = 20 + (possession_home * 0.7)
        x       = np.random.normal(loc=mean_x, scale=15, size=800)
        y       = np.random.normal(loc=40, scale=20*(1+abs(sentiment)), size=800)
        x       = np.clip(x, 0, 120)
        y       = np.clip(y, 0, 80)

        pitch.kdeplot(x, y, ax=ax, fill=True, levels=100, cmap='hot')
        ax.set_title(f"Field Pressure Heatmap: {time_label} min",
                     color='white', fontsize=14)
        plt.show()

    # ─────────────────────────────────────────────────────────────────────────
    # MAIN RUNNER
    # ─────────────────────────────────────────────────────────────────────────
    def run(self):
        print("\n" + "═"*80)
        print("  ⚽  SOCCER COMBINED TACTICAL & PREDICTIVE PIPELINE  ⚽")
        print("═"*80)

        combined_data = self.load_and_combine()
        self.identify_teams(combined_data)
        t1, t2 = self.teams

        cur_h, cur_a  = 0, 0
        prev_pbp_text = None
        prev_sentiment = None

        for i in range(4):
            time_label = f"{i*30}-{(i+1)*30}"
            start_s    = i * 1800
            end_s      = (i + 1) * 1800

            block = [s for s in combined_data if start_s <= s['start'] < end_s]
            if not block:
                continue

            raw_text = " ".join(s['text'] for s in block)
            print(f"\n{'─'*60}")
            print(f"  SEGMENT [{time_label} min]")
            print(f"{'─'*60}")

            # INJECT COMPRESSION HERE
            compressed_text = self.compress_transcript(raw_text)

            # ── 2a: NARRATIVE ──────────────────────────────────────────────
            print("\n  ▶ Generating narrative summary...")

            # PASS THE COMPRESSED TEXT INSTEAD OF RAW TEXT
            narr_data, narr_fallback, narr_lat = self.generate_narrative(
                compressed_text, t1, t2, cur_h, cur_a)
            pbp_text = narr_data.get('play_by_play', '')

            print("  ── Verification: Narrative ──")
            self.verifier.log_check(self.verifier.record_latency("narrative", narr_lat))
            self.verifier.log_check(self.verifier.record_call_result(
                "narrative", not narr_fallback, narr_fallback))
            self.verifier.log_check(self.verifier.check_schema(
                narr_data, NARRATIVE_SCHEMA, "narrative"))
            self.verifier.log_check(self.verifier.check_completeness(
                narr_data,
                ["play_by_play","new_scorers","fouls_mentioned",
                 "cards_mentioned","topic_label","crowd_intensity","referee_involvement"],
                "narrative"))
            self.verifier.log_check(self.verifier.check_narrative_coherence(
                narr_data, time_label))

            # ── 3: NLP FEATURES ───────────────────────────────────────────
            sentiment, new_h, new_a = self.extract_nlp_features(pbp_text, cur_h, cur_a)
            if i > 0:
                # Calculate the influence percentage for the print statement
                carryover_weight = prev_sentiment * 0.7
                tilt_effect = carryover_weight * 15 # Scale to match your win_prob math

                print(f"\n   [TEMPORAL FLOW] Influence from {i*30-30}-{i*30}m Segment:")
                print(f"   • Previous Momentum Charge : {prev_sentiment:+.2f}")
                print(f"   • Narrative Buffer Applied  : {carryover_weight:+.2f} (70% Decay Weight)")
                print(f"   • Probability Tilt          : This historical 'charge' is physically shifting the ")
                print(f"                                 current {time_label}m Win Probability by {tilt_effect:+.1f}%.")
            else:
                print(f"\n   [TEMPORAL START] Initializing Baseline: No previous match history to carry over.")

            print("  ── Verification: Score & Sentiment ──")
            self.verifier.log_check(self.verifier.check_score_integrity(
                cur_h, cur_a, new_h, new_a, time_label))
            self.verifier.log_check(self.verifier.check_hallucination_proxy(
                narr_data, raw_text, time_label))
            self.verifier.log_check(self.verifier.check_sentiment_calibration(
                sentiment,
                narr_data.get("crowd_intensity", 0.5),
                time_label))

            # Update scores
            prev_h, prev_a = cur_h, cur_a
            cur_h,  cur_a  = new_h, new_a

            # ── TEMPORAL DYNAMICS ─────────────────────────────────────────
            if prev_pbp_text is not None and prev_sentiment is not None:
                sentiment_delta   = sentiment - prev_sentiment
                tactical_sim      = self.nlp(pbp_text).similarity(self.nlp(prev_pbp_text))
            else:
                sentiment_delta   = 0.0
                tactical_sim      = 0.0

            print("  ── Verification: Cross-Segment Drift ──")
            self.verifier.log_check(self.verifier.check_cross_segment_drift(
                narr_data.get("topic_label","performance"), sentiment, time_label))

            prev_pbp_text  = pbp_text
            prev_sentiment = sentiment

            # ── LABELS ────────────────────────────────────────────────────
            sentiment_label  = self.categorize_sentiment(sentiment)
            crowd_label      = self.categorize_intensity(narr_data.get("crowd_intensity", 0.5))
            referee_label    = self.categorize_referee(narr_data.get("referee_involvement", 0.5))
            momentum_label   = self.categorize_momentum(sentiment_delta)
            tactical_state   = self.categorize_tactics(tactical_sim)
            possession_home  = max(20.0, min(80.0, 50.0 + sentiment * 10))

            # ── CONSOLE PRINT ─────────────────────────────────────────────
            print(f"\n  📋 Score     : {t1} {cur_h} - {cur_a} {t2}")
            print(f"  📊 Sentiment : {sentiment:.3f}  |  Label: {sentiment_label}")
            print(f"  🏃 Momentum  : {momentum_label}  |  Tactical: {tactical_state}")
            print(f"  👥 Crowd     : {crowd_label}  |  Referee: {referee_label}")
            print(f"  ⚽ Scorers   : {narr_data.get('new_scorers', [])}")
            print(f"  📝 Narrative :\n     {pbp_text[:300]}...")

            # ── 2b: COACHING ADVICE ───────────────────────────────────────
            print("\n  ▶ Generating coaching advice...")
            coach_data, coach_fallback, coach_lat = self.get_coach_advice(pbp_text, t1, t2)

            print("  ── Verification: Coaching ──")
            self.verifier.log_check(self.verifier.record_latency("coaching", coach_lat))
            self.verifier.log_check(self.verifier.record_call_result(
                "coaching", not coach_fallback, coach_fallback))
            self.verifier.log_check(self.verifier.check_schema(
                coach_data, COACH_SCHEMA, "coaching"))
            self.verifier.log_check(self.verifier.check_completeness(
                coach_data, ["home_team_analysis","away_team_analysis"], "coaching"))
            self.verifier.log_check(self.verifier.check_coaching_specificity(
                coach_data, time_label))

            home_coach = coach_data.get('home_team_analysis', {})
            away_coach = coach_data.get('away_team_analysis', {})
            print(f"\n  🎯 COACHING – {t1}:")
            print(f"     Alert   : {home_coach.get('substitution_alert')}")
            print(f"     Tactics : {home_coach.get('tactical_shift')}")
            print(f"     Advice  : {home_coach.get('coach_recommendation')}")
            print(f"\n  🎯 COACHING – {t2}:")
            print(f"     Alert   : {away_coach.get('substitution_alert')}")
            print(f"     Tactics : {away_coach.get('tactical_shift')}")
            print(f"     Advice  : {away_coach.get('coach_recommendation')}")

            # ── DATASET ROW ───────────────────────────────────────────────
            row = {
                "time_segment":      time_label,
                "home_score":        cur_h, "away_score": cur_a,
                "goals_diff":        cur_h - cur_a,
                "sentiment_score":   sentiment, "sentiment_label": sentiment_label,
                "sentiment_delta":   sentiment_delta, "momentum_label": momentum_label,
                "tactical_similarity": tactical_sim, "tactical_state": tactical_state,
                "crowd_label":       crowd_label, "referee_label": referee_label,
                "topic_label":       narr_data.get("topic_label","performance"),
                "fouls":             narr_data.get("fouls_mentioned", 0),
                "cards":             narr_data.get("cards_mentioned", 0),
                "possession_home":   possession_home,
                "target_match_outcome": (
                    "Home_Win" if cur_h > cur_a else
                    ("Draw" if cur_h == cur_a else "Away_Win")
                )
            }
            self.dataset.append(row)

            # ── 2c: PREDICTION VALIDATION ─────────────────────────────────
            win_prob   = max(0.01, min(0.99,
                0.33 + (row['goals_diff'] * 0.25) + (row['sentiment_score'] * 0.1)))
            pred_label = ("Home_Win" if win_prob > 0.5 else
                          ("Away_Win" if win_prob < 0.3 else "Draw"))

            print(f"\n  ▶ Validating model prediction: {pred_label} ({win_prob:.2f})")
            val_data, val_fallback, val_lat = self.validate_prediction(row, pred_label, win_prob)

            print("  ── Verification: Prediction Validation ──")
            self.verifier.log_check(self.verifier.record_latency("validation", val_lat))
            self.verifier.log_check(self.verifier.record_call_result(
                "validation", not val_fallback, val_fallback))
            self.verifier.log_check(self.verifier.check_schema(
                val_data, VALIDATION_SCHEMA, "validation"))
            self.verifier.log_check(self.verifier.check_validation_output(val_data, time_label))

            print(f"     Consistency : {val_data.get('consistency_score', 'N/A')}")
            print(f"     Explanation : {val_data.get('explanation', 'N/A')}")

            # ── 5: HEATMAP ────────────────────────────────────────────────
            self.generate_heatmap(possession_home, sentiment, time_label)

        # ── FINAL DATASET ─────────────────────────────────────────────────
        df = pd.DataFrame(self.dataset)
        print("\n" + "═"*80)
        print("  FINAL TIMELINE DATASET")
        print("═"*80)
        print(df.to_string())

        # ── VERIFICATION REPORT ───────────────────────────────────────────
        self.verifier.print_summary_report()
        self.verifier.export_log_csv("verification_log.csv")

        return df


# ─────────────────────────────────────────────────────────────────────────────
# ENTRY POINT
# ─────────────────────────────────────────────────────────────────────────────
if __name__ == "__main__":
    pipeline = SoccerCombinedPipeline('/content/1_asr.json', '/content/2_asr.json')
    pipeline.run()
```

```python
import json
import os
import re
import time
import pandas as pd
import numpy as np
import nltk
import spacy
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from mplsoccer import Pitch
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from groq import Groq
from jsonschema import validate, ValidationError
from collections import defaultdict

# ─────────────────────────────────────────────────────────────────────────────
# DEPENDENCIES: pip install groq spacy mplsoccer jsonschema nltk pandas numpy
#               python -m spacy download en_core_web_md
# ─────────────────────────────────────────────────────────────────────────────

try:
    nltk.data.find('sentiment/vader_lexicon.zip')
except LookupError:
    nltk.download('vader_lexicon', quiet=True)

# ─────────────────────────────────────────────────────────────────────────────
# CONFIGURATION
# ─────────────────────────────────────────────────────────────────────────────
API_KEY     = "<api_key>"
client      = Groq(api_key=API_KEY)
GROQ_MODEL  = "llama-3.3-70b-versatile"

# ─────────────────────────────────────────────────────────────────────────────
# JSON SCHEMAS FOR STRICT VALIDATION
# ─────────────────────────────────────────────────────────────────────────────
NARRATIVE_SCHEMA = {
    "type": "object",
    "required": ["play_by_play", "new_scorers", "fouls_mentioned",
                 "cards_mentioned", "topic_label", "crowd_intensity",
                 "referee_involvement"],
    "properties": {
        "play_by_play":         {"type": "string", "minLength": 30},
        "new_scorers":          {"type": "array",  "items": {"type": "string"}},
        "fouls_mentioned":      {"type": "integer","minimum": 0, "maximum": 50},
        "cards_mentioned":      {"type": "integer","minimum": 0, "maximum": 20},
        "topic_label":          {"type": "string",
                                 "enum": ["tactical","emotional","referee","performance"]},
        "crowd_intensity":      {"type": "number", "minimum": 0.0, "maximum": 1.0},
        "referee_involvement":  {"type": "number", "minimum": 0.0, "maximum": 1.0},
    },
    "additionalProperties": True,
}

COACH_SCHEMA = {
    "type": "object",
    "required": ["home_team_analysis", "away_team_analysis"],
    "properties": {
        "home_team_analysis": {
            "type": "object",
            "required": ["substitution_alert", "tactical_shift", "coach_recommendation"],
            "properties": {
                "substitution_alert":    {"type": ["string", "null"]},
                "tactical_shift":        {"type": "string", "minLength": 5},
                "coach_recommendation":  {"type": "string", "minLength": 10},
            }
        },
        "away_team_analysis": {
            "type": "object",
            "required": ["substitution_alert", "tactical_shift", "coach_recommendation"],
            "properties": {
                "substitution_alert":    {"type": ["string", "null"]},
                "tactical_shift":        {"type": "string", "minLength": 5},
                "coach_recommendation":  {"type": "string", "minLength": 10},
            }
        }
    }
}

VALIDATION_SCHEMA = {
    "type": "object",
    "required": ["consistency_score", "explanation"],
    "properties": {
        "consistency_score": {"type": "number", "minimum": 0.0, "maximum": 1.0},
        "explanation":       {"type": "string", "minLength": 10},
    }
}

SCORESTATE_RE = re.compile(r"SCORESTATE:\s*.*?(\d+)\s*-\s*(\d+)", re.IGNORECASE)

# ─────────────────────────────────────────────────────────────────────────────
# ██  VERIFICATION ENGINE
# ─────────────────────────────────────────────────────────────────────────────
class LLMOutputVerifier:
    """
    Runs multi-dimensional quality checks on every LLM output.

    Metric families
    ───────────────
    1. Schema Compliance       – JSON structure & type validation
    2. Field Completeness      – Required keys present & non-empty
    3. Semantic Coherence      – SCORESTATE present, topic valid, etc.
    4. Score Integrity         – Non-decreasing, realistic jumps
    5. Hallucination Proxy     – Ungrounded goal spikes, implausible values
    6. Sentiment Calibration   – VADER vs LLM crowd-intensity correlation
    7. Latency Tracking        – API response time
    8. Retry / Fallback Rate   – How often the LLM fails
    9. Cross-Segment Drift     – Topic & narrative consistency over time
    10. Coaching Specificity   – Tactical advice richness score
    """

    def __init__(self):
        self.log: list[dict] = []                # Per-call records
        self.segment_history: list[dict] = []    # Cross-segment state
        self.summary_stats: dict = defaultdict(list)

    # ── helpers ─────────────────────────────────────────────────────────────
    @staticmethod
    def _word_count(text: str) -> int:
        return len(text.split()) if text else 0

    @staticmethod
    def _schema_check(data: dict, schema: dict) -> tuple[bool, str]:
        try:
            validate(instance=data, schema=schema)
            return True, "OK"
        except ValidationError as e:
            return False, e.message

    # ── METRIC 1: Schema Compliance ─────────────────────────────────────────
    def check_schema(self, data: dict, schema: dict, call_name: str) -> dict:
        passed, msg = self._schema_check(data, schema)
        return {"metric": "schema_compliance", "call": call_name,
                "passed": passed, "detail": msg}

    # ── METRIC 2: Field Completeness ────────────────────────────────────────
    def check_completeness(self, data: dict, required_keys: list,
                           call_name: str) -> dict:
        missing, empty = [], []
        for k in required_keys:
            if k not in data:
                missing.append(k)
            elif data[k] in (None, "", [], {}):
                empty.append(k)
        passed = not missing and not empty
        detail = (f"Missing: {missing}; Empty: {empty}") if not passed else "All fields present"
        return {"metric": "field_completeness", "call": call_name,
                "passed": passed, "detail": detail,
                "missing_keys": missing, "empty_keys": empty}

    # ── METRIC 3: Semantic Coherence ────────────────────────────────────────
    def check_narrative_coherence(self, narrative_data: dict,
                                  time_label: str) -> dict:
        pbp       = narrative_data.get("play_by_play", "")
        has_score = bool(SCORESTATE_RE.search(pbp))
        wc        = self._word_count(pbp)
        topic_ok  = narrative_data.get("topic_label") in \
                    ["tactical", "emotional", "referee", "performance"]

        issues = []
        if not has_score:  issues.append("SCORESTATE tag missing from play_by_play")
        if wc < 40:        issues.append(f"play_by_play too short ({wc} words)")
        if not topic_ok:   issues.append(f"Invalid topic_label: {narrative_data.get('topic_label')}")

        return {"metric": "semantic_coherence", "segment": time_label,
                "passed": not issues, "scorestate_found": has_score,
                "word_count": wc, "topic_valid": topic_ok, "issues": issues}

    # ── METRIC 4: Score Integrity ────────────────────────────────────────────
    def check_score_integrity(self, prev_h: int, prev_a: int,
                              new_h: int, new_a: int,
                              time_label: str) -> dict:
        issues = []
        if new_h < prev_h:
            issues.append(f"Home score decreased: {prev_h} → {new_h}")
        if new_a < prev_a:
            issues.append(f"Away score decreased: {prev_a} → {new_a}")
        delta_h = new_h - prev_h
        delta_a = new_a - prev_a
        if delta_h > 4:
            issues.append(f"Unrealistic home goal burst (+{delta_h} in 30 min)")
        if delta_a > 4:
            issues.append(f"Unrealistic away goal burst (+{delta_a} in 30 min)")
        if new_h > 15 or new_a > 15:
            issues.append(f"Implausible total score: {new_h}-{new_a}")

        return {"metric": "score_integrity", "segment": time_label,
                "passed": not issues, "prev_score": f"{prev_h}-{prev_a}",
                "new_score": f"{new_h}-{new_a}",
                "delta_home": delta_h, "delta_away": delta_a, "issues": issues}

    # ── METRIC 5: Hallucination Proxy ───────────────────────────────────────
    def check_hallucination_proxy(self, narrative_data: dict,
                                  raw_text: str, time_label: str) -> dict:
        """
        Flags potential hallucinations by checking if claimed scorers
        appear anywhere in the raw transcript.
        """
        scorers   = narrative_data.get("new_scorers", [])
        raw_lower = raw_text.lower()
        phantom   = [s for s in scorers
                     if s and s.lower() not in raw_lower and s not in ("None","")]
        ratio     = len(phantom) / max(len(scorers), 1)
        passed    = ratio < 0.5  # Allow < 50% unverified names

        return {"metric": "hallucination_proxy", "segment": time_label,
                "passed": passed, "claimed_scorers": scorers,
                "unverified_scorers": phantom,
                "phantom_ratio": round(ratio, 3)}

    # ── METRIC 6: Sentiment Calibration ─────────────────────────────────────
    def check_sentiment_calibration(self, vader_compound: float,
                                    llm_crowd_intensity: float,
                                    time_label: str) -> dict:
        """
        VADER compound ∈ [-1,1]; crowd_intensity ∈ [0,1].
        They should positively correlate. We flag if they diverge sharply.
        """
        normalised_vader = (vader_compound + 1) / 2   # → [0,1]
        gap              = abs(normalised_vader - llm_crowd_intensity)
        passed           = gap < 0.45

        return {"metric": "sentiment_calibration", "segment": time_label,
                "passed": passed,
                "vader_compound": round(vader_compound, 4),
                "llm_crowd_intensity": round(llm_crowd_intensity, 4),
                "normalised_vader": round(normalised_vader, 4),
                "calibration_gap": round(gap, 4),
                "detail": "OK" if passed else f"Large gap ({gap:.2f}) – LLM may be over/under-stating crowd"}

    # ── METRIC 7: Latency Tracking ───────────────────────────────────────────
    def record_latency(self, call_name: str, elapsed_s: float) -> dict:
        return {"metric": "api_latency_s", "call": call_name,
                "elapsed": round(elapsed_s, 3),
                "slow": elapsed_s > 15}

    # ── METRIC 8: Retry / Fallback Rate ─────────────────────────────────────
    def record_call_result(self, call_name: str, success: bool,
                           fallback_used: bool) -> dict:
        self.summary_stats["successes"].append(int(success))
        self.summary_stats["fallbacks"].append(int(fallback_used))
        return {"metric": "call_success", "call": call_name,
                "success": success, "fallback_used": fallback_used}

    # ── METRIC 9: Cross-Segment Drift ───────────────────────────────────────
    def check_cross_segment_drift(self, topic: str, sentiment: float,
                                  time_label: str) -> dict:
        """Detects abrupt, suspicious topic flips between segments."""
        result = {"metric": "cross_segment_drift", "segment": time_label,
                  "topic": topic, "sentiment": round(sentiment, 4),
                  "passed": True, "detail": "First segment"}

        if self.segment_history:
            prev = self.segment_history[-1]
            topic_flip = (prev["topic"] != topic)
            sent_jump  = abs(sentiment - prev["sentiment"])

            suspicious = topic_flip and sent_jump > 0.7
            result.update({
                "prev_topic": prev["topic"], "topic_flipped": topic_flip,
                "sentiment_jump": round(sent_jump, 4),
                "passed": not suspicious,
                "detail": ("Suspicious abrupt topic+sentiment swing"
                           if suspicious else "Drift within acceptable range")
            })

        self.segment_history.append({"topic": topic, "sentiment": sentiment})
        return result

    # ── METRIC 10: Coaching Specificity ─────────────────────────────────────
    def check_coaching_specificity(self, coach_data: dict,
                                   time_label: str) -> dict:
        """
        Scores coaching output richness:
        – Uses a named player for sub alert (+2)
        – Tactical shift has a formation number e.g. 4-3-3 (+2)
        – Recommendation ≥ 15 words (+1)
        """
        scores, details = {}, {}
        for side in ("home_team_analysis", "away_team_analysis"):
            d     = coach_data.get(side, {})
            score = 0
            sub   = d.get("substitution_alert", "") or ""
            tact  = d.get("tactical_shift", "") or ""
            rec   = d.get("coach_recommendation", "") or ""

            if sub.lower() not in ("none", "n/a", "", "null") and len(sub) > 2:
                score += 2
            if re.search(r'\d-\d-\d', tact):
                score += 2
            if self._word_count(rec) >= 15:
                score += 1

            scores[side]  = score
            details[side] = {"sub_alert_named": score >= 2,
                             "formation_cited": score >= 4,
                             "rec_length_ok": self._word_count(rec) >= 15}

        total   = sum(scores.values())
        passed  = total >= 3  # At least half the possible 10 points
        return {"metric": "coaching_specificity", "segment": time_label,
                "passed": passed, "total_score": total,
                "max_score": 10, "breakdown": details}

    # ── METRIC 11: Consistency Score Sanity ─────────────────────────────────
    def check_validation_output(self, val_data: dict, time_label: str) -> dict:
        c   = val_data.get("consistency_score", -1)
        exp = val_data.get("explanation", "")
        issues = []
        if not (0.0 <= c <= 1.0):
            issues.append(f"consistency_score out of range: {c}")
        if self._word_count(exp) < 5:
            issues.append("Explanation too brief")
        return {"metric": "validation_output_sanity", "segment": time_label,
                "passed": not issues, "consistency_score": c,
                "explanation_word_count": self._word_count(exp),
                "issues": issues}

    # ── Log & append ────────────────────────────────────────────────────────
    def log_check(self, result: dict):
        self.log.append(result)
        status = "✅ PASS" if result.get("passed", True) else "❌ FAIL"
        print(f"   [{status}] {result['metric']}"
              + (f" | {result.get('detail','')}" if result.get('detail') else "")
              + (f" | issues: {result.get('issues','')}" if result.get('issues') else ""))

    # ── Final Report ─────────────────────────────────────────────────────────
    def print_summary_report(self):
        print("\n" + "═"*80)
        print("  🔬  LLM OUTPUT VERIFICATION SUMMARY REPORT")
        print("═"*80)

        total  = len(self.log)
        passed = sum(1 for r in self.log if r.get("passed", True))
        failed = total - passed

        print(f"\n  Total checks   : {total}")
        print(f"  ✅ Passed       : {passed}  ({passed/max(total,1)*100:.1f}%)")
        print(f"  ❌ Failed       : {failed}  ({failed/max(total,1)*100:.1f}%)")

        # Group by metric family
        family_map: dict[str, list] = defaultdict(list)
        for r in self.log:
            family_map[r["metric"]].append(r)

        print("\n  ─── Per-Metric Breakdown ───")
        for family, records in family_map.items():
            p = sum(1 for r in records if r.get("passed", True))
            print(f"  {family:<35} {p}/{len(records)} passed")

        # Latency stats
        latencies = [r["elapsed"] for r in self.log if r.get("metric") == "api_latency_s"]
        if latencies:
            print(f"\n  ─── Latency Stats ───")
            print(f"  Mean  : {np.mean(latencies):.2f}s")
            print(f"  Max   : {np.max(latencies):.2f}s")
            print(f"  Slow calls (>15s): {sum(1 for l in latencies if l>15)}")

        # Retry / Fallback
        fb = self.summary_stats.get("fallbacks", [])
        if fb:
            print(f"\n  ─── Fallback Rate ───")
            print(f"  Fallback used: {sum(fb)}/{len(fb)} calls ({sum(fb)/len(fb)*100:.1f}%)")

        # Failed checks detail
        failures = [r for r in self.log if not r.get("passed", True)]
        if failures:
            print("\n  ─── Failed Checks Detail ───")
            for f in failures:
                seg  = f.get("segment", f.get("call", "?"))
                issl = f.get("issues", f.get("detail", ""))
                print(f"  ● [{f['metric']}] @ {seg}: {issl}")

        print("\n" + "═"*80 + "\n")

    # ── Export ───────────────────────────────────────────────────────────────
    def export_log_csv(self, path: str = "verification_log.csv"):
        pd.DataFrame(self.log).to_csv(path, index=False)
        print(f"  Verification log exported → {path}")


# ─────────────────────────────────────────────────────────────────────────────
# ██  COMBINED PIPELINE
# ─────────────────────────────────────────────────────────────────────────────
class SoccerCombinedPipeline:

    def __init__(self, h1_path: str, h2_path: str):
        self.h1_path = h1_path
        self.h2_path = h2_path

        print("Loading NLP Models (VADER & spaCy)...")
        self.sia = SentimentIntensityAnalyzer()
        self.nlp = spacy.load("en_core_web_md")

        self.teams: list[str] = []
        self.dataset: list[dict] = []
        self.verifier = LLMOutputVerifier()

    # ─────────────────────────────────────────────────────────────────────────
    # STEP 1: DATA INGESTION
    # ─────────────────────────────────────────────────────────────────────────
    def load_and_combine(self) -> list[dict]:
        def get_segs(path):
            if not os.path.exists(path):
                raise FileNotFoundError(f"File not found: {path}")
            with open(path, 'r', encoding='utf-8') as f:
                data = json.load(f)
            segs = data.get("segments", {})
            return sorted(segs.items(), key=lambda x: int(x[0]))

        h1_segs = get_segs(self.h1_path)
        h2_segs = get_segs(self.h2_path)

        combined = [{"start": v[0], "text": v[2].replace('\n', ' ').strip()}
                    for _, v in h1_segs]
        combined += [{"start": v[0] + 2700, "text": v[2].replace('\n', ' ').strip()}
                     for _, v in h2_segs]
        return combined

    def identify_teams(self, combined_data: list[dict]):
        print("Extracting team names via Groq LLM...")
        intro  = " ".join(s['text'] for s in combined_data if s['start'] <= 600)
        prompt = (f"Identify the TWO competing football clubs.\n"
                  f"Return JSON: {{\"home_team\": \"<String>\", \"away_team\": \"<String>\"}}\n"
                  f"Transcript: \"{intro}\"")

        fallback_used = False
        t0 = time.time()
        try:
            resp = client.chat.completions.create(
                messages=[{"role": "user", "content": prompt}],
                model=GROQ_MODEL,
                response_format={"type": "json_object"},
            )
            data = json.loads(resp.choices[0].message.content)
            self.teams = [data.get("home_team", "Home Team"),
                          data.get("away_team",  "Away Team")]
        except Exception as e:
            print(f"  Team extraction failed: {e}")
            self.teams = ["Home Team", "Away Team"]
            fallback_used = True

        elapsed = time.time() - t0
        self.verifier.log_check(self.verifier.record_latency("identify_teams", elapsed))
        self.verifier.log_check(
            self.verifier.record_call_result("identify_teams", not fallback_used, fallback_used))
        print(f"  Match: {self.teams[0]} (Home) vs {self.teams[1]} (Away)")
    def compress_transcript(self, raw_text: str) -> str:
        """
        Compresses raw ASR transcript by filtering out neutral, non-action sentences.
        Retains sentences with high emotional sentiment or key tactical football terms.
        """
        # Fast sentence splitting using regex (looks for periods, exclamation marks, etc.)
        sentences = re.split(r'(?<=[.!?])\s+', raw_text)

        # High-value vocabulary for soccer analytics
        keywords = {
            'goal', 'score', 'shot', 'save', 'foul', 'card', 'yellow',
            'red', 'penalty', 'corner', 'sub', 'replace', 'miss',
            'offside', 'var', 'cross', 'header', 'tackle', 'press', 'injury'
        }

        compressed_sentences = []

        for sent in sentences:
            if not sent.strip():
                continue

            sent_lower = sent.lower()

            # Rule 1: Does it contain a tactical keyword?
            has_keyword = any(kw in sent_lower for kw in keywords)

            # Rule 2: Is the commentator highly emotional? (VADER compound score)
            # Threshold of 0.4 catches strong excitement or sharp disappointment
            sentiment_score = self.sia.polarity_scores(sent)['compound']
            is_emotional = abs(sentiment_score) > 0.4

            # Keep the sentence if it passes either rule
            if has_keyword or is_emotional:
                compressed_sentences.append(sent.strip())

        compressed_text = " ".join(compressed_sentences)

        # Calculate and print compression stats
        orig_words = len(raw_text.split())
        comp_words = len(compressed_text.split())
        ratio = (1 - (comp_words / max(orig_words, 1))) * 100

        print(f"  ✂️ [COMPRESSION] Reduced from {orig_words} to {comp_words} words (-{ratio:.1f}%)")

        # Safety Fallback: If the filter was too aggressive and wiped out the text,
        # revert to the original raw text to prevent pipeline failure.
        if comp_words < 30 and orig_words > 100:
            print("  ⚠️ [COMPRESSION] Warning: Too aggressive. Bypassing compression.")
            return raw_text

        return compressed_text

    # ─────────────────────────────────────────────────────────────────────────
    # STEP 2a: NARRATIVE GENERATION (from Pipeline 1)
    # ─────────────────────────────────────────────────────────────────────────
    def generate_narrative(self, raw_text: str, t1: str, t2: str,
                           cur_h: int, cur_a: int) -> tuple[dict, bool, float]:
        prompt = f"""
        You are a sports journalist. Read this 30-min transcript between {t1} (Home) and {t2} (Away).
        Write a chronological play_by_play summary. Track goals, near misses, fouls, and cards.
        MANDATORY ENDING: SCORESTATE: {t1} X - Y {t2}
        (Replace X,Y with cumulative integer scores, adding to previous {cur_h}-{cur_a})

        Return JSON:
        {{
            "play_by_play": "<summary ending with SCORESTATE>",
            "new_scorers": ["<string>"],
            "fouls_mentioned": 0, "cards_mentioned": 0,
            "topic_label": "tactical",
            "crowd_intensity": 0.5, "referee_involvement": 0.5
        }}
        Transcript: "{raw_text}"
        """
        fallback_used = False
        t0 = time.time()
        try:
            resp = client.chat.completions.create(
                messages=[{"role": "user", "content": prompt}],
                model=GROQ_MODEL,
                response_format={"type": "json_object"},
            )
            data = json.loads(resp.choices[0].message.content)
        except Exception as e:
            print(f"  Narrative API Error: {e}")
            data = {
                "play_by_play": f"Error processing segment. SCORESTATE: {t1} {cur_h} - {cur_a} {t2}",
                "new_scorers": [], "fouls_mentioned": 0, "cards_mentioned": 0,
                "topic_label": "performance", "crowd_intensity": 0.5, "referee_involvement": 0.5
            }
            fallback_used = True
        return data, fallback_used, time.time() - t0

    # ─────────────────────────────────────────────────────────────────────────
    # STEP 2b: COACHING ADVICE (from Pipeline 2)
    # ─────────────────────────────────────────────────────────────────────────
    def get_coach_advice(self, narrative: str, home: str, away: str) -> tuple[dict, bool, float]:
        prompt = f"""
        You are an elite football manager. Analyze this 30-min match segment:
        "{narrative}"

        Provide tactical recommendations and substitution alerts for {home} and {away}.
        Identify specific players making errors or showing fatigue where mentioned.

        Return strictly JSON:
        {{
            "home_team_analysis": {{
                "substitution_alert": "<player name or null>",
                "tactical_shift": "<e.g. Change to 4-5-1>",
                "coach_recommendation": "<brief strategic advice>"
            }},
            "away_team_analysis": {{
                "substitution_alert": "<player name or null>",
                "tactical_shift": "<e.g. High press required>",
                "coach_recommendation": "<brief strategic advice>"
            }}
        }}
        """
        fallback_used = False
        t0 = time.time()
        try:
            resp = client.chat.completions.create(
                messages=[{"role": "user", "content": prompt}],
                model=GROQ_MODEL,
                response_format={"type": "json_object"},
            )
            data = json.loads(resp.choices[0].message.content)
        except Exception as e:
            print(f"  Coaching API Error: {e}")
            data = {"home_team_analysis": {"substitution_alert": None,
                                           "tactical_shift": "N/A",
                                           "coach_recommendation": "N/A"},
                    "away_team_analysis": {"substitution_alert": None,
                                           "tactical_shift": "N/A",
                                           "coach_recommendation": "N/A"}}
            fallback_used = True
        return data, fallback_used, time.time() - t0

    # ─────────────────────────────────────────────────────────────────────────
    # STEP 2c: PREDICTION VALIDATION (from Pipeline 1)
    # ─────────────────────────────────────────────────────────────────────────
    def validate_prediction(self, features: dict, prediction: str,
                            win_prob: float) -> tuple[dict, bool, float]:
        prompt = f"""
        Given these 30-min football segment features:
        - Topic: {features['topic_label']}
        - Sentiment Label: {features['sentiment_label']}
        - Momentum Shift: {features['momentum_label']}
        - Tactical State: {features['tactical_state']}
        - Events: {features['goals_diff']} Goal Difference, {features['fouls']} Fouls, {features['cards']} Cards

        An XGBoost model predicted: {prediction} (Home Win Probability: {win_prob:.2f}).
        Is this prediction consistent with the narrative context?

        Return JSON:
        {{
            "consistency_score": 0.0,
            "explanation": "<brief explanation>"
        }}
        """
        fallback_used = False
        t0 = time.time()
        try:
            resp = client.chat.completions.create(
                messages=[{"role": "user", "content": prompt}],
                model=GROQ_MODEL,
                response_format={"type": "json_object"},
            )
            data = json.loads(resp.choices[0].message.content)
        except Exception as e:
            print(f"  Validation API Error: {e}")
            data = {"consistency_score": 0.5, "explanation": "Validation failed."}
            fallback_used = True
        return data, fallback_used, time.time() - t0

    # ─────────────────────────────────────────────────────────────────────────
    # STEP 3: DETERMINISTIC NLP EXTRACTION
    # ─────────────────────────────────────────────────────────────────────────
    def extract_nlp_features(self, text: str,
                             fallback_h: int, fallback_a: int) -> tuple[float, int, int]:
        sentiment = self.sia.polarity_scores(text)['compound']
        match     = SCORESTATE_RE.search(text)
        if match:
            return sentiment, int(match.group(1)), int(match.group(2))
        return sentiment, fallback_h, fallback_a

    # ─────────────────────────────────────────────────────────────────────────
    # STEP 4: CATEGORICAL LABELS
    # ─────────────────────────────────────────────────────────────────────────
    @staticmethod
    def categorize_sentiment(s: float) -> str:
        if s >= 0.5:  return "Highly Exciting"
        if s >= 0.05: return "Positive / Active"
        if s > -0.05: return "Neutral / Balanced"
        return "Tense / Negative"

    @staticmethod
    def categorize_intensity(s: float) -> str:
        if s >= 0.7: return "Roaring / Electric"
        if s >= 0.4: return "Moderate / Murmuring"
        return "Quiet"

    @staticmethod
    def categorize_referee(s: float) -> str:
        if s >= 0.6: return "Strict / Highly Involved"
        if s >= 0.3: return "Moderate"
        return "Letting the Game Flow"

    @staticmethod
    def categorize_momentum(delta: float) -> str:
        if delta > 0.3:  return "Home Surge"
        if delta < -0.3: return "Away Surge"
        return "Stable"

    @staticmethod
    def categorize_tactics(sim: float) -> str:
        if sim == 0.0: return "N/A (Match Start)"
        if sim > 0.85: return "Stagnant / Consistent"
        return "Major Tactical Shift"

    # ─────────────────────────────────────────────────────────────────────────
    # STEP 5: SPATIAL HEATMAP (from Pipeline 2)
    # ─────────────────────────────────────────────────────────────────────────
    def generate_heatmap(self, possession_home: float, sentiment: float,
                         time_label: str):
        pitch = Pitch(pitch_type='statsbomb', pitch_color='#1a1a1a', line_color='#c7d5cc')
        fig, ax = pitch.draw(figsize=(10, 6))

        mean_x  = 20 + (possession_home * 0.7)
        x       = np.random.normal(loc=mean_x, scale=15, size=800)
        y       = np.random.normal(loc=40, scale=20*(1+abs(sentiment)), size=800)
        x       = np.clip(x, 0, 120)
        y       = np.clip(y, 0, 80)

        pitch.kdeplot(x, y, ax=ax, fill=True, levels=100, cmap='hot')
        ax.set_title(f"Field Pressure Heatmap: {time_label} min",
                     color='white', fontsize=14)
        plt.show()

    # ─────────────────────────────────────────────────────────────────────────
    # MAIN RUNNER
    # ─────────────────────────────────────────────────────────────────────────
    def run(self):
        print("\n" + "═"*80)
        print("  ⚽  SOCCER COMBINED TACTICAL & PREDICTIVE PIPELINE  ⚽")
        print("═"*80)

        combined_data = self.load_and_combine()
        self.identify_teams(combined_data)
        t1, t2 = self.teams

        cur_h, cur_a  = 0, 0
        prev_pbp_text = None
        prev_sentiment = None

        for i in range(4):
            time_label = f"{i*30}-{(i+1)*30}"
            start_s    = i * 1800
            end_s      = (i + 1) * 1800

            block = [s for s in combined_data if start_s <= s['start'] < end_s]
            if not block:
                continue

            raw_text = " ".join(s['text'] for s in block)
            print(f"\n{'─'*60}")
            print(f"  SEGMENT [{time_label} min]")
            print(f"{'─'*60}")

            # INJECT COMPRESSION HERE
            compressed_text = self.compress_transcript(raw_text)

            # ── 2a: NARRATIVE ──────────────────────────────────────────────
            print("\n  ▶ Generating narrative summary...")

            # PASS THE COMPRESSED TEXT INSTEAD OF RAW TEXT
            narr_data, narr_fallback, narr_lat = self.generate_narrative(
                compressed_text, t1, t2, cur_h, cur_a)
            pbp_text = narr_data.get('play_by_play', '')

            print("  ── Verification: Narrative ──")
            self.verifier.log_check(self.verifier.record_latency("narrative", narr_lat))
            self.verifier.log_check(self.verifier.record_call_result(
                "narrative", not narr_fallback, narr_fallback))
            self.verifier.log_check(self.verifier.check_schema(
                narr_data, NARRATIVE_SCHEMA, "narrative"))
            self.verifier.log_check(self.verifier.check_completeness(
                narr_data,
                ["play_by_play","new_scorers","fouls_mentioned",
                 "cards_mentioned","topic_label","crowd_intensity","referee_involvement"],
                "narrative"))
            self.verifier.log_check(self.verifier.check_narrative_coherence(
                narr_data, time_label))

            # ── 3: NLP FEATURES ───────────────────────────────────────────
            sentiment, new_h, new_a = self.extract_nlp_features(pbp_text, cur_h, cur_a)
            if i > 0:
                # Calculate the influence percentage for the print statement
                carryover_weight = prev_sentiment * 0.7
                tilt_effect = carryover_weight * 15 # Scale to match your win_prob math

                print(f"\n   [TEMPORAL FLOW] Influence from {i*30-30}-{i*30}m Segment:")
                print(f"   • Previous Momentum Charge : {prev_sentiment:+.2f}")
                print(f"   • Narrative Buffer Applied  : {carryover_weight:+.2f} (70% Decay Weight)")
                print(f"   • Probability Tilt          : This historical 'charge' is physically shifting the ")
                print(f"                                 current {time_label}m Win Probability by {tilt_effect:+.1f}%.")
            else:
                print(f"\n   [TEMPORAL START] Initializing Baseline: No previous match history to carry over.")

            print("  ── Verification: Score & Sentiment ──")
            self.verifier.log_check(self.verifier.check_score_integrity(
                cur_h, cur_a, new_h, new_a, time_label))
            self.verifier.log_check(self.verifier.check_hallucination_proxy(
                narr_data, raw_text, time_label))
            self.verifier.log_check(self.verifier.check_sentiment_calibration(
                sentiment,
                narr_data.get("crowd_intensity", 0.5),
                time_label))

            # Update scores
            prev_h, prev_a = cur_h, cur_a
            cur_h,  cur_a  = new_h, new_a

            # ── TEMPORAL DYNAMICS ─────────────────────────────────────────
            if prev_pbp_text is not None and prev_sentiment is not None:
                sentiment_delta   = sentiment - prev_sentiment
                tactical_sim      = self.nlp(pbp_text).similarity(self.nlp(prev_pbp_text))
            else:
                sentiment_delta   = 0.0
                tactical_sim      = 0.0

            print("  ── Verification: Cross-Segment Drift ──")
            self.verifier.log_check(self.verifier.check_cross_segment_drift(
                narr_data.get("topic_label","performance"), sentiment, time_label))

            prev_pbp_text  = pbp_text
            prev_sentiment = sentiment

            # ── LABELS ────────────────────────────────────────────────────
            sentiment_label  = self.categorize_sentiment(sentiment)
            crowd_label      = self.categorize_intensity(narr_data.get("crowd_intensity", 0.5))
            referee_label    = self.categorize_referee(narr_data.get("referee_involvement", 0.5))
            momentum_label   = self.categorize_momentum(sentiment_delta)
            tactical_state   = self.categorize_tactics(tactical_sim)
            possession_home  = max(20.0, min(80.0, 50.0 + sentiment * 10))

            # ── CONSOLE PRINT ─────────────────────────────────────────────
            print(f"\n  📋 Score     : {t1} {cur_h} - {cur_a} {t2}")
            print(f"  📊 Sentiment : {sentiment:.3f}  |  Label: {sentiment_label}")
            print(f"  🏃 Momentum  : {momentum_label}  |  Tactical: {tactical_state}")
            print(f"  👥 Crowd     : {crowd_label}  |  Referee: {referee_label}")
            print(f"  ⚽ Scorers   : {narr_data.get('new_scorers', [])}")
            print(f"  📝 Narrative :\n     {pbp_text[:300]}...")

            # ── 2b: COACHING ADVICE ───────────────────────────────────────
            print("\n  ▶ Generating coaching advice...")
            coach_data, coach_fallback, coach_lat = self.get_coach_advice(pbp_text, t1, t2)

            print("  ── Verification: Coaching ──")
            self.verifier.log_check(self.verifier.record_latency("coaching", coach_lat))
            self.verifier.log_check(self.verifier.record_call_result(
                "coaching", not coach_fallback, coach_fallback))
            self.verifier.log_check(self.verifier.check_schema(
                coach_data, COACH_SCHEMA, "coaching"))
            self.verifier.log_check(self.verifier.check_completeness(
                coach_data, ["home_team_analysis","away_team_analysis"], "coaching"))
            self.verifier.log_check(self.verifier.check_coaching_specificity(
                coach_data, time_label))

            home_coach = coach_data.get('home_team_analysis', {})
            away_coach = coach_data.get('away_team_analysis', {})
            print(f"\n  🎯 COACHING – {t1}:")
            print(f"     Alert   : {home_coach.get('substitution_alert')}")
            print(f"     Tactics : {home_coach.get('tactical_shift')}")
            print(f"     Advice  : {home_coach.get('coach_recommendation')}")
            print(f"\n  🎯 COACHING – {t2}:")
            print(f"     Alert   : {away_coach.get('substitution_alert')}")
            print(f"     Tactics : {away_coach.get('tactical_shift')}")
            print(f"     Advice  : {away_coach.get('coach_recommendation')}")

            # ── DATASET ROW ───────────────────────────────────────────────
            row = {
                "time_segment":      time_label,
                "home_score":        cur_h, "away_score": cur_a,
                "goals_diff":        cur_h - cur_a,
                "sentiment_score":   sentiment, "sentiment_label": sentiment_label,
                "sentiment_delta":   sentiment_delta, "momentum_label": momentum_label,
                "tactical_similarity": tactical_sim, "tactical_state": tactical_state,
                "crowd_label":       crowd_label, "referee_label": referee_label,
                "topic_label":       narr_data.get("topic_label","performance"),
                "fouls":             narr_data.get("fouls_mentioned", 0),
                "cards":             narr_data.get("cards_mentioned", 0),
                "possession_home":   possession_home,
                "target_match_outcome": (
                    "Home_Win" if cur_h > cur_a else
                    ("Draw" if cur_h == cur_a else "Away_Win")
                )
            }
            self.dataset.append(row)

            # ── 2c: PREDICTION VALIDATION ─────────────────────────────────
            win_prob   = max(0.01, min(0.99,
                0.33 + (row['goals_diff'] * 0.25) + (row['sentiment_score'] * 0.1)))
            pred_label = ("Home_Win" if win_prob > 0.5 else
                          ("Away_Win" if win_prob < 0.3 else "Draw"))

            print(f"\n  ▶ Validating model prediction: {pred_label} ({win_prob:.2f})")
            val_data, val_fallback, val_lat = self.validate_prediction(row, pred_label, win_prob)

            print("  ── Verification: Prediction Validation ──")
            self.verifier.log_check(self.verifier.record_latency("validation", val_lat))
            self.verifier.log_check(self.verifier.record_call_result(
                "validation", not val_fallback, val_fallback))
            self.verifier.log_check(self.verifier.check_schema(
                val_data, VALIDATION_SCHEMA, "validation"))
            self.verifier.log_check(self.verifier.check_validation_output(val_data, time_label))

            print(f"     Consistency : {val_data.get('consistency_score', 'N/A')}")
            print(f"     Explanation : {val_data.get('explanation', 'N/A')}")

            # ── 5: HEATMAP ────────────────────────────────────────────────
            self.generate_heatmap(possession_home, sentiment, time_label)

        # ── FINAL DATASET ─────────────────────────────────────────────────
        df = pd.DataFrame(self.dataset)
        print("\n" + "═"*80)
        print("  FINAL TIMELINE DATASET")
        print("═"*80)
        print(df.to_string())

        # ── VERIFICATION REPORT ───────────────────────────────────────────
        self.verifier.print_summary_report()
        self.verifier.export_log_csv("verification_log.csv")

        return df


# ─────────────────────────────────────────────────────────────────────────────
# ENTRY POINT
# ─────────────────────────────────────────────────────────────────────────────
if __name__ == "__main__":
    pipeline = SoccerCombinedPipeline('/content/1_asr (3).json', '/content/2_asr (3).json')
    pipeline.run()
```
